# GPS tracks to line segments: Connecting two dots

This is the first in a potential series of notebooks to try and figure out how to take GPS tracks from Runkeeper and transform them into individual line segments in a new dataset. It is rather straight forward to make a single line feature from the full set of points. It seems to be more complicated and less common to take points and create individual segments for consecutive points. In this series, I'm going to try and build from simply creating a single line segment from the first two points in my GPS datasets to creating individual segments for each pair of points, to eventually doing calculations of eleveation change/speed/distance covered over each segment.

## Import libraries

The most important 3 libraries here are pandas, geopandas, and folium.

````python
%matplotlib inline
import os
import pandas as pd
import geopandas as gpd
import folium
````

## Some useful function for later on

### Plot markers from a GeoDataFrame to folium map

````python
def add_markers(mapobj, gdf):
    coords = []
    for i, row in gdf.iterrows():
        coords.append([row.geometry.y, row.geometry.x])
    for coord in coords:
        folium.CircleMarker(location = coord,
                            radius = 2.5, 
                            fill = True,
                            fill_color = '#F50057',
                            fill_opacity = 0.75,
                            color = 'whitesmoke',
                            weight = 0.5).add_to(mapobj)
    return mapobj
````

### Get the x,y coordinates for a given row of point GeoDataFrame

````python
def get_coords(gdf, i, geometry = 'geometry'):
    track_geom = gdf.loc[i, geometry]
    x, y = [track_geom.x, track_geom.y]
    return x, y
````

## Load GPS points into a GeoDataFrame

Using geopanda's, we'll load a shapefile of track points into a GeoDataFrame. 

````python
gps = 'timemanager_test_data.shp'
tracks = gpd.GeoDataFrame.from_file(gps)
tracks.head()
````

Looking at the data, there are several fields that we don't need. Let's make a copy of our dataset containing only attributes info about collection sequence (`track_se_`), elevation (`ele`), time (`time`), and the recorded location for each track point.

````python
tracks = tracks[['track_se_1', 'ele', 'time', 'geometry']].copy()
tracks.head()
````

The geometry column is a shapely `POINT Z()` object, which means in addition to longitude and latitude, it also include the elevation (Z) value. We don't need that elevation value at the moment. It's not really a problem that it's there either. We can still plot track points just fine by passing a folium map object and our tracks GeoDataFrame to the `add_markers()` function defined earlier.

````python
f = folium.Figure(height = 400)
m = folium.Map([35.792809, -78.675724], zoom_start = 15, tiles='Cartodb dark_matter')
m.add_to(f)

add_markers(m, tracks)
````

## Connect two dots

Cool, so we are able to plot the GeoDataFrame of track points to a map. But how can we connect those point? We need to create a line segment using the coordinates of one track point as the start and the coordinates of the next track point as the segments end. What information do we have to help us do that? Let's try to connect the first two track points to find out.

Ultimately what we're trying to create are several very simple instances of a geometry type called a LineString. We can construct LineString geometries using, at a minimum, two pairs of x,y coordinates. A LineString geometry looks like this:

```Python
LineString([(x0, y0), (x1, y1)])
```

Our tracks data have the information we need to form sequential two-point LineStrings. X,Y coordinates and a value for where each track falls in the sequence of the tracks collected.

````python
tracks[['track_se_1','geometry']].head(2)
````

We can access the coordinates for first and second records using the `get_coord()` functions we created earlier.

````python
x0, y0 = get_coords(tracks, 0)
print(x0, y0)
````

````python
x1, y1 = get_coords(tracks, 1)
print(x1, y1)
````

Create a LineString from `x0, y0` and `x1, y1`.

````python
from shapely.geometry import LineString
````

````python
line = LineString([(x0, y0), (x1, y1)])
line
````

````python
data = {'id': [0], 'geometry': [line]}
df = pd.DataFrame(data, columns = ['id', 'geometry'])
df
````

````python
from geopandas import GeoDataFrame
````

````python
crs = {'init': 'epsg:4326'}
gdf = GeoDataFrame(df, crs=crs)
gdf.plot()
````

````python
folium.GeoJson(gdf).add_to(m)
m
````
